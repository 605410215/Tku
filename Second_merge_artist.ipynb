```python
import pandas as pd
import numpy as np
import time as tm
```

```python
df2=pd.read_csv('song.json',sep='\t')
df2.rename(columns={'id': 'song_id'}, inplace=True)
df2
```

```python
size=0
start_time=tm.time()
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='1'+chr(ii)+chr(j)
        df=pd.read_csv(file_name,sep='\t')
        df=df.merge(df2,on='song_id', how='left')
        df.drop(['song_id'], axis=1, inplace=True)
        df.to_csv(file_name,sep='\t',encoding='utf-8',index=False)
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break
end_time=tm.time()-start_time
print ('Data_size:%d Process_time:%d'%(size,end_time))
```
